# BUSINESS UNDERSTANDING

##Study Case

Dalam balapan Formula One, pit stop merupakan salah satu strategi penting yang dapat menentukan kemenangan tim. Pit stop yang terlalu lama dapat menyebabkan pembalap kehilangan posisi, sedangkan pit stop yang cepat dan tepat waktu dapat meningkatkan peluang menang.

##Business Objective

Business objective-nya adalah untuk membantu *racing team* meningkatkan performa dan efisiensi strategi pit stop dengan menganalisis *history* pit stop agar dapat mencapai waktu pit stop secepat mungkin dan mendapatkan *placement* terbaik mereka setelah balapan berakhir.

Data ini berisi data pit stop dari balapan Formula One. Data tersebut mencatat informasi tentang pembalap, tim, lap saat pit stop dilakukan, jumlah pit stop, serta durasi waktu pit stop selama balapan.

# Assess Situation

Tim sering kesulitan menentukan strategi pit stop terbaik. Banyak hal yang harus dipertimbangkan, seperti berapa kali pit stop, kapan waktu yang tepat untuk pit stop, dan berapa lama durasinya. Karena itu, diperlukan analisis data pit stop sebelumnya untuk melihat pola dan menentukan strategi yang paling efisien.
